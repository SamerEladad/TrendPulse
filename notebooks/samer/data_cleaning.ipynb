##  Join 12 Datasets into one big dataset

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to the folder containing the datasets
folder_path = "/Users/samer/Documents/github_repos/TrendPulse/data/raw/Amazon Kaggle Datasets"

# List of filenames to be joined
filenames = [
    "Jewellery.csv",
    "All_Electronics.csv",
    "Kitchen_and_Home_Appliances.csv",
    "Travel_Accessories.csv",
    "Beauty_and_Grooming.csv",
    "Car_Accessories.csv",
    "Fitness_Accessories.csv",
    "Car_Electronics.csv",
    "Home_Improvement.csv",
    "All_Pet_Supplies.csv",
    "Toys_and_Games.csv",
    "Wallets.csv"
]

# Read and concatenate all files
all_data = pd.concat([pd.read_csv(os.path.join(folder_path, filename)) for filename in filenames], ignore_index=True)

# Export the combined data to a new CSV file
output_file = os.path.join(folder_path, "Combined_Amazon_Dataset_Raw.csv")
all_data.to_csv(output_file, index=False)

print(f"Combined dataset created and saved as: {output_file}")
```

```python
# Read the combined dataset
df  = pd.read_csv(output_file)
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Print the columns of the dataframe
print(df.columns)
```

```python
# Plotting the number of different main categories
plt.figure(figsize=(10, 5))
df['main_category'].value_counts(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Number of Different Main Categories')
plt.xlabel('Main Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting the number of different sub categories
plt.figure(figsize=(10, 5))
df['sub_category'].value_counts(ascending=False).plot(kind='bar', color='lightgreen')
plt.title('Number of Different Sub Categories')
plt.xlabel('Sub Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting the number of items per main category
plt.figure(figsize=(10, 5))
df.groupby('main_category').size().sort_values(ascending=False).plot(kind='bar', color='coral')
plt.title('Number of Items per Main Category')
plt.xlabel('Main Category')
plt.ylabel('Number of Items')
plt.xticks(rotation=90)
plt.show()
```

```python
# Remove rows where 'ratings' is 'FREE' or 'Get'
df_cleaned = df_cleaned[~df_cleaned['ratings'].isin(['FREE', 'Get'])]

# Replace NaN values in 'ratings' with 0.0
df_cleaned['ratings'] = df_cleaned['ratings'].fillna(0.0).astype(float)

# Verify the changes
print(df_cleaned['ratings'].unique())
```

```python
# Replace NaN values in 'ratings' with 0.0 and convert to float
df_cleaned['ratings'] = df_cleaned['ratings'].fillna(0.0).astype(float)

# Print the unique values of the 'ratings' column again
unique_ratings = df_cleaned['ratings'].unique()
print(list(unique_ratings))
```

```python
# Change the 'ratings' data type to float
df_cleaned['ratings'] = df_cleaned['ratings'].astype(float)
```

```python
# Remove the '₹' symbol from all elements in the 'discount_price' column
df_cleaned['discount_price'] = df_cleaned['discount_price'].str.replace('₹', '', regex=False).str.strip()

# Remove the ',' symbol from all elements in the 'discount_price' column
df_cleaned['discount_price'] = df_cleaned['discount_price'].str.replace(',', '', regex=False).str.strip()

# Convert the column to a numeric type (float) after removing the symbol
df_cleaned['discount_price'] = pd.to_numeric(df_cleaned['discount_price'], errors='coerce')

# Verify the changes
print(df_cleaned['discount_price'].head())
```

```python
# Remove the '₹' symbol from all elements in the 'actual_price' column
df_cleaned['actual_price'] = df_cleaned['actual_price'].str.replace('₹', '', regex=False).str.strip()

# Remove the ',' symbol from all elements in the 'actual_price' column
df_cleaned['actual_price'] = df_cleaned['actual_price'].str.replace(',', '', regex=False).str.strip()

# onvert the column to a numeric type (float) after removing the symbol
df_cleaned['actual_price'] = pd.to_numeric(df_cleaned['actual_price'], errors='coerce')

# Verify the changes
print(df_cleaned['actual_price'].head())
```

```python
# Remove the ',' symbol from all elements in the 'no_of_ratings' column
df_cleaned['no_of_ratings'] = df_cleaned['no_of_ratings'].str.replace(',', '', regex=False).str.strip()

# onvert the column to a numeric type (float) after removing the symbol
df_cleaned['no_of_ratings'] = pd.to_numeric(df_cleaned['no_of_ratings'], errors='coerce')

# Verify the changes
print(df_cleaned['no_of_ratings'].head())
```
